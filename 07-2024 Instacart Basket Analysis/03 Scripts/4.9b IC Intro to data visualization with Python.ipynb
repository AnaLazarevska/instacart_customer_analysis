## 05. Create sample subset and line chart for prices at various times of the day

```python
# create the sample

np.random.seed(4)
dev = np.random.rand(len(ords_prods_cust_merged)) <= 0.7
```

```python
dev
```

```python
# split the dataframe in two samples

big = ords_prods_cust_merged[dev]
```

```python
small = ords_prods_cust_merged[~dev]
```

```python
# check if the ratio of big and small samples is 70/30, first check the number of rows in the dataframe

len(ords_prods_cust_merged)
```

```python
# check the number of rows in big and small samples

len(big) + len(small)
```

```python
# get two columns from the small sample

df_2 = small[['order_hour_of_day','prices']]
```

```python
# create line chart from two columns in small sample

line = sns.lineplot(data = df_2, x = 'order_hour_of_day',y = 'prices')
```

## Answer to question 2.5

The line chart above shows that the orders made just after midnight at more expensive than the orders made just before midnight. For the rest of the day, the prices of the orders are around 10 to 13 dollars.

## 06. Create line chart for age and number of dependents

```python
# get two columns from the small sample

df_3 = small[['age','fam_status']]
```

```python
# create line chart from two columns in small sample

line_1 = sns.lineplot(data = df_3, x = 'fam_status',y = 'age')
```

The line chart shows that most of the customers in their 50s are married, in their 70s are either widowed or divorced, in their 40s single, and in their 20s living with parents.

## 07. Create scatterplot

```python
# create scatterplot

scat = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_cust_merged)
```

## Answer to question 2.7

The schatterplot shows equal distribution of low income for all age grops. The income is higher for the older population, and lower for the younger population.
