## 04. Create loyalty flag for existing customers using the transform() and loc() functions

```python
# create max_order column with results from transform() function on the âorder_numberâ column grouped by user_id

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(10)
```

```python
# create flag for loyalty labels

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
ords_prods_merge.head()
```

## 05. Identify spending habits of the three types of customers

```python
# find mean prices grouped by customer category

ords_prods_merge.groupby('loyalty_flag')['prices'].mean()
```

### How do prices differ between customer categories?

Even though there are fewer new customers than both loyatly and regular customers, they purchase more expensive products on average than the other two types of customers. The products they buy are 1 dollar more expensive than the products that the regular customers buy. The loyalty customers purchase the cheapest products, spending 10 dollars on average.

## 06. Identify average spending price for each customers

```python
# create spending_flag column for each user based on the average orders price

ords_prods_merge['spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# create spending flag function for each user based on average price

ords_prods_merge.loc[ords_prods_merge['spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts()
```

```python
ords_prods_merge.tail()
```

## 07. Identify user's ordering frequency

```python
# Create order frequency flag that marks the regularity of a userâs ordering behavior according to the median in the âdays_since_prior_orderâ column

ords_prods_merge['frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# create frequency flag function for each user based on median days_since_prior_order

ords_prods_merge.loc[ords_prods_merge['frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['frequency'] > 10) & (ords_prods_merge['frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts()
```

```python
ords_prods_merge.head()
```
