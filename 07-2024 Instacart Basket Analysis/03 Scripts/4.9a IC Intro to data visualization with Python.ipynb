## 04. Consistency checks

```python
# find mixed data types

for col in cust.columns.tolist():
  weird = (cust[[col]].map(type) != cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (cust[weird]) > 0:
    print (col)
```

```python
# check data type of the identified mixed data column

cust['first_name'].dtype
```

```python
# check for null values in column with mixed data types

cust.isnull().any()
```

```python
# find missing values

cust_nan = cust[cust['first_name'].isnull() == True]
```

```python
cust_nan
```

## Answer to question 1.5: address missing values

Since the missing values NaN are all in the first_name column, they play little part in the analysis, but the corresponding rows still hold data that is valuable for the analysis (every row that has missing first name has valuable information about the person's purchasing power). Therefore, I will only change the data type to 'str' instead of dropping those values. 

```python
# address mixed data types and missing values

cust['first_name'] = cust['first_name'].astype('str')
```
