## 03. Create variables and functions

```python
# create subset from df_ords_prods

df = ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
# create function

def price_label_func(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) & (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return np.nan
```

```python
# apply the function

df['price_range'] = df.apply(price_label_func, axis=1)
```

```python
# display number of values in price_range categories

df['price_range'].value_counts(dropna = False)
```

```python
# check highest price

df['prices'].max()
```

## 06. For-loop

```python
# count values in orders_days_of_week

ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# create for-loop function

result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create new column for the for-loop result

ords_prods_merge['busiest_day'] = result
```

## 08. Create new column for busiest_days

```python
# calculate value count for days

day_counts = ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
day_counts
```

```python
# create for-loop function

updated_result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if (value == 0 or value == 1):
    updated_result.append("Two busiest days")
  elif (value == 4 or value == 3):
    updated_result.append("Two least busy days")
  else:
    updated_result.append("Regularly busy")
```

```python
# create new column for the for-loop result

ords_prods_merge['busiest_days'] = updated_result
```

## 10. Create new column for busiest_period_of_day

```python
# calculate value count for hours

hour_counts = ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
hour_counts
```

```python
# create for-loop function

hours_result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if (value >= 9 and value <= 16):
    hours_result.append("Most orders")
  elif (value >= 2 and value <= 5):
    hours_result.append("Fewest orders")
  else:
    hours_result.append("Average orders")
```

```python
# create new column for the for-loop result

ords_prods_merge['busiest_period_of_day'] = hours_result
```

```python
# get value count for the new column

ords_prods_merge['busiest_period_of_day'].value_counts()
```
